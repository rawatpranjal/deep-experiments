```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

folder_path = '/econ_share/home/pp712/deep-experiments/data'
excluded_estimators = []

def update_scores(df, r):
    df_filtered = df[~df["Estimator"].isin(excluded_estimators)]
    if not df_filtered.empty:
        for _, row in df_filtered.iterrows():
            estimator = row["Estimator"]
            bias_score_complex = r * (-row["DGP Complex: Abs Bias (%)"])
            bias_score_simple = r * (-row["DGP Simple: Abs Bias (%)"])
            var_score_complex = row["DGP Complex: Var Dec (%)"]
            var_score_simple = row["DGP Simple: Var Dec (%)"]
            total_score = bias_score_complex + bias_score_simple + var_score_complex + var_score_simple
            scores[estimator] = scores.get(estimator, 0) + total_score

best_estimators = {}
second_best_estimators = {}
r_values = list(np.linspace(0.0001, 100, 100))
change_points = []

previous_best = None

for r in r_values:
    scores = {}
    for filename in os.listdir(folder_path):
        if filename.endswith('.csv'):
            file_path = os.path.join(folder_path, filename)
            df = pd.read_csv(file_path, skiprows=1, names=["Estimator", "DGP Complex: Abs Bias (%)", "DGP Complex: Var Dec (%)", "DGP Simple: Abs Bias (%)", "DGP Simple: Var Dec (%)"])
            update_scores(df, r)
    if scores:
        leaderboard = pd.DataFrame(list(scores.items()), columns=["Estimator", "Total Score"]).sort_values(by="Total Score", ascending=False)
        best_estimator = leaderboard.iloc[0, 0]
        second_best_estimator = leaderboard.iloc[1, 0]
        best_estimators[r] = best_estimator
        second_best_estimators[r] = second_best_estimator
        
        # Track changes in the best estimator
        if previous_best is None:
            previous_best = best_estimator
        elif best_estimator != previous_best:
            change_points.append(r)
            previous_best = best_estimator

# Print the r values at which the best estimator changes
print("Best estimator changes at the following r values:")
for cp in change_points:
    print(f"r = {cp:.4f}")

# Plot the best estimator
plt.figure(figsize=(10, 5))
plt.plot(r_values, [best_estimators[r] for r in r_values], marker='o', label='Best Estimator')
plt.xlabel('r (Bias to Variance Conversion Factor)')
plt.ylabel('Best Estimator')
plt.title('Best Estimator vs r')
plt.xticks([])  # Removes x-axis tick labels
plt.grid(False)  # Removes grid
plt.legend()
plt.show()

# Plot the second-best estimator
plt.figure(figsize=(10, 5))
plt.plot(r_values, [second_best_estimators[r] for r in r_values], marker='o', label='Second Best Estimator', color='orange')
plt.xlabel('r (Bias to Variance Conversion Factor)')
plt.ylabel('Second Best Estimator')
plt.title('Second Best Estimator vs r')
plt.xticks([])  # Removes x-axis tick labels
plt.grid(False)  # Removes grid
plt.legend()
plt.show()
```
